```python
import cv2
import numpy as np
import importlib
count=0

def get_output_layers(net):
    layer_names = net.getLayerNames()
    output_layers = [layer_names[i[0] - 1] for i in net.getUnconnectedOutLayers()]
    return output_layers

def draw_prediction(img, class_id, confidence, x, y, w, h):
    global img
    print(confidence)
    global axle_detected
    label = str(classes[class_id])
    #print(label,":",confidence)
    #print( )
    color = COLORS[class_id]
    print(label)
    if label == "Helmet":
        #print("hi")
        #crop_img = img[y-5:y+h+5, x-5:x+w+5]
        axle_detected = True

with open("yolov3-tiny-obj.names", 'r') as f:
    classes = [line.strip() for line in f.readlines()]
    print(len(classes))

COLORS = np.random.uniform(0, 255, size=(len(classes), 3))
net = cv2.dnn.readNet("yolov3-tiny-obj_1000.weights", "yolov3-tiny-obj.cfg")
    
for i in range(2):
    frame = cv2.imread("test.jpg")
    
    cv2.imshow("frame",frame)
    #cv2.waitKey(0)
    #frame = cv2.resize(frame,(400,300))
    if frame is None:
        continue
    print(frame.shape)
    #frame = frame[700:1400,800:1600]
    #print(frame.shape)
#     Car_detected = False
#     crop_img_2 = None
#     crop_img = None
    axle_detected = False
    scale = 0.003
    print(frame.shape)
    
    #frame = cv2.resize(image,(600,300))
    Width = frame.shape[1]
    Height = frame.shape[0]
    blob = cv2.dnn.blobFromImage(frame, scale, (416,416), (0,0,0), True, crop=False)
    net.setInput(blob)
    outs = net.forward(get_output_layers(net))
    class_ids = []
    confidences = []
    boxes = []
    conf_threshold = 0.3
    nms_threshold = 0.3
    for out in outs:
        for detection in out:
            scores = detection[5:]
            class_id = np.argmax(scores)
            confidence = scores[class_id]
            if confidence > 0.3:
                center_x = int(detection[0] * Width)
                center_y = int(detection[1] * Height)
                w = int(detection[2] * Width)
                h = int(detection[3] * Height)
                x = center_x - w / 2
                y = center_y - h / 2
                class_ids.append(class_id)
                confidences.append(float(confidence))
                boxes.append([x, y, w, h])
    indices = cv2.dnn.NMSBoxes(boxes, confidences, conf_threshold, nms_threshold)
    #print(indices)
    for i in indices:
        i = i[0]
        box = boxes[i]
        x = box[0]
        y = box[1]
        w = box[2]
        h = box[3]
        draw_prediction(frame, class_ids[i], confidences[i], round(x), round(y), round(w), round(h))
        
    if axle_detected == True:
        print("detected")
        cv2.imshow("Helmet",frame)
        #count+=1
        
```
